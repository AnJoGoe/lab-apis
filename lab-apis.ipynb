1. **List Your Favorite Artists**:

```python
# 1. Make a list of your three favorite artists and store it in a variable named artists.
artists = ["Iron Maiden", "Florence + The Machine", "Stromae"]
```

2. **Fetch Top Tracks**:
    
    - This function should accept an artist's name and return the name of the first 5 top tracks by that artist.
    - Use the function `get_top_tracks` to get the first 5 top tracks for each artist in your `artists` list and store the results in a new list named `top_tracks_list`.

```python
# Write a function named `get_top_tracks`.

def get_top_tracks(artist: str):
    """
    function takes a list of artist names and returns the first 5 top tracks by that artist
    """
    results = sp.search(q=artist, limit=5)
    
    top_5 = []
    
    for i in range(5):
       top_5.append(results["tracks"]["items"][i]["name"]) 
    
    return top_5
```

```python
get_top_tracks(artists[0])
```

```python
#Use the function `get_top_tracks` to get the first 5 top tracks for each artist in your `artists` list and store the results in a new list named `top_tracks_list`.

top_tracks_list = [get_top_tracks(x) for x in artists] 
top_tracks_list
```

3. **Discover Related Artists**:

```python
def find_related_artists(artist: str):
    """
    returns the first 5 artists related to a provided artist
    """
    results = sp.search(q=artist, limit=1, type="artist")
    
    # get artist id
    artist_id = results["artists"]["items"][0]["id"] 
    
    related_artists = sp.artist_related_artists(artist_id)
    
    related_list = []
    for i in range(5):
        related_list.append(related_artists["artists"][i]["name"])
    
    return related_list
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        #if artist_name not in artists_list:  # This ensures each artist is added only once
        artists_list.append(artist_name)

print(len(artists_list))
```

   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
tracks = sp.playlist_tracks('7G99lH3FE5E3ZtHho7N1mF')
tracks["items"][0]["track"]["artists"][0]["name"]# track artist
```

```python
def track_tastic (playlist_id):
    tracks_list = []
    tracks = sp.playlist_tracks(playlist_id)

    try:
        for track in tracks['items']:
            tracks_list.append(track['track']['name'])
    except:
        pass
    
    return tracks_list[:10]

#display(len(tracks_list[:10]))


def get_artists_for_song(track_id):
    try:
        # Retrieve information about the track
        track_info = sp.track(track_id)
        
        # Extract the list of artists for the track
        artists_list = [artist['name'] for artist in track_info['artists']]
        
        return artists_list
    except Exception as e:
        print("An error occurred:", e)
        return []
    

    
    
def artistic_flair (playlist_id):
    # get 10 tracks for playlist id
    track_list = track_tastic(playlist_id)
    
    
    # get track id
    
    track_id_dict = {}
    
    for track_name in track_list:
        
        if track_id_dict.get(track_name) is None:
            results = sp.search(q=track_name, type='track', limit=1)
            track_id_dict[track_name] = results['tracks']['items'][0]['id']

    track_artist_dict = {}
    
    for key,value in track_id_dict.items():
        
        if track_artist_dict.get(value) is None:
            track_artist_dict[key] = get_artists_for_song(value)
            
    return  track_artist_dict
    
        
    
    
    
    
    
```
